# **3.0 Visualizations**
---

```python
plt.figure(figsize=(16, 6))
sns.histplot(df['PRICE_CURRENT'], bins=50, kde=True)
plt.title('Distribution of Current Product Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.scatterplot(x='PRICE_RETAIL', y='PRICE_CURRENT', data=df)
plt.title('Retail Price vs Current Price')
plt.xlabel('Retail Price ($)')
plt.ylabel('Current Price ($)')
plt.show()

```

```python
top_brands = df['BRAND'].value_counts().head(10)
plt.figure(figsize=(16 , 6))
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.title('Top 10 Brands by Number of Products')
plt.xlabel('Brand')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_price_department = df.groupby('DEPARTMENT')['PRICE_CURRENT'].mean().sort_values()
plt.figure(figsize=(16, 6))
sns.barplot(x=avg_price_department.index, y=avg_price_department.values)
plt.title('Average Current Price by Department')
plt.xlabel('Department')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
product_count_location = df['SHIPPING_LOCATION'].value_counts()
plt.figure(figsize=(16, 6))
sns.barplot(x=product_count_location.index, y=product_count_location.values)
plt.title('Product Count by Shipping Location')
plt.xlabel('Shipping Location')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.boxplot(x='DEPARTMENT', y='PRICE_CURRENT', data=df)
plt.title('Price Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Current Price ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_price_category = df.groupby('CATEGORY')['PRICE_CURRENT'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(16 , 6))
sns.barplot(x=avg_price_category.index, y=avg_price_category.values)
plt.title('Average Price by Category')
plt.xlabel('Category')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=90)
plt.show()
```

```python
product_count_subcategory = df['SUBCATEGORY'].value_counts().head(20)
plt.figure(figsize=(16 , 6))
sns.barplot(x=product_count_subcategory.index, y=product_count_subcategory.values)
plt.title('Top 20 Subcategories by Product Count')
plt.xlabel('Subcategory')
plt.ylabel('Number of Products')
plt.xticks(rotation=90)
plt.show()
```

```python
top_10_brands = df['BRAND'].value_counts().head(10).index
avg_price_top_brands = df[df['BRAND'].isin(top_10_brands)].groupby('BRAND')['PRICE_CURRENT'].mean()
plt.figure(figsize=(16 , 6))
sns.barplot(x=avg_price_top_brands.index, y=avg_price_top_brands.values)
plt.title('Average Price by Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_retail_price_category = df.groupby('CATEGORY')['PRICE_RETAIL'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(16 , 6))
sns.barplot(x=avg_retail_price_category.index, y=avg_retail_price_category.values)
plt.title('Top 10 Categories by Average Retail Price')
plt.xlabel('Category')
plt.ylabel('Average Retail Price ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_current_price_subcategory = df.groupby('SUBCATEGORY')['PRICE_CURRENT'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(16 , 6))
sns.barplot(x=avg_current_price_subcategory.index, y=avg_current_price_subcategory.values)
plt.title('Top 10 Subcategories by Average Current Price')
plt.xlabel('Subcategory')
plt.ylabel('Average Current Price ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
most_expensive_products = df[['PRODUCT_NAME', 'PRICE_CURRENT']].sort_values(by='PRICE_CURRENT', ascending=False).head(10)
plt.figure(figsize=(16 , 6))
sns.barplot(x=most_expensive_products['PRODUCT_NAME'], y=most_expensive_products['PRICE_CURRENT'])
plt.title('Top 10 Most Expensive Products')
plt.xlabel('Product Name')
plt.ylabel('Current Price ($)')
plt.xticks(rotation=12)
plt.show()
```

```python
avg_retail_price_brand = df.groupby('BRAND')['PRICE_RETAIL'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(16, 6))
sns.barplot(x=avg_retail_price_brand.index, y=avg_retail_price_brand.values)
plt.title('Top 10 Brands by Average Retail Price')
plt.xlabel('Brand')
plt.ylabel('Average Retail Price ($)')
plt.xticks(rotation=45)
plt.show()
```

# **4.0 Data Insights**
---


### **4.1 What is the most common shipping location ?**

```python
most_common_shipping_location = df['SHIPPING_LOCATION'].mode()[0]
print(f"The most common shipping location is {most_common_shipping_location}.")
```

### **4.2 What is the average price of products ?**

```python
average_current_price = df['PRICE_CURRENT'].mean()
print(f"The average current price of products is ${average_current_price:.2f}.")
```

### **4.3 What is the most average retail price of the products ?**

```python
average_retail_price = df['PRICE_RETAIL'].mean()
print(f"The average retail price of products is ${average_retail_price:.2f}.")
```

### **4.4 What percentage of products is on sale ?**

```python
products_on_sale = df[df['PRICE_RETAIL'] > df['PRICE_CURRENT']].shape[0]
total_products = df.shape[0]
percentage_on_sale = (products_on_sale / total_products) * 100
print(f"{percentage_on_sale:.2f}% of products are currently on sale.")
```

### **4.5 How many unique brands are there in dataSet ?**

```python
unique_brands = df['BRAND'].nunique()
print(f"There are {unique_brands} unique brands in the dataset.")
```

### **4.6 What are the number of products per department ?**

```python
products_per_department = df['DEPARTMENT'].value_counts()
print("Number of products per department:")
print(products_per_department)
```

### **4.7 Which is the most expensive product and what's it's price ?**

```python
most_expensive_product = df.loc[df['PRICE_CURRENT'].idxmax()]
print(f"The most expensive product is '{most_expensive_product['PRODUCT_NAME']}' with a current price of ${most_expensive_product['PRICE_CURRENT']:.2f}.")
```

### **4.8 Which is the most inexpensive product and what is it's price ?**

```python
cheapest_product = df.loc[df['PRICE_CURRENT'].idxmin()]
print(f"The cheapest product is '{cheapest_product['PRODUCT_NAME']}' with a current price of ${cheapest_product['PRICE_CURRENT']:.2f}.")
```

### **4.9 What is the total of unique products avilable ?**

```python
total_products = df['SKU'].nunique()
print(f"There are a total of {total_products} unique products available.")
```

### **4.10 What is the average product size ?**

```python
average_product_size = df['PRODUCT_SIZE'].mode()[0]
print(f"The average product size is {average_product_size:.2f}.")
```

### **4.11 What is the most common category ?**

```python
most_common_category = df['CATEGORY'].mode()[0]
print(f"The most common product category is {most_common_category}.")
```

### **4.12 What is the total sales value ?**

```python
total_sales_value = df['PRICE_CURRENT'].sum()
print(f"The total sales value (current price) is ${total_sales_value:.2f}.")
```

### **4.13 What is the total retail value ?**

```python
total_retail_value = df['PRICE_RETAIL'].sum()
print(f"The total retail value is ${total_retail_value:.2f}.")
```

### **4.14 What is the number of products per category ?**

```python
products_per_subcategory = df['SUBCATEGORY'].value_counts()
print("Number of products per subcategory:")
print(products_per_subcategory)
```

### **4.15 What is the average difference between retail and current price ?**

```python
df['PRICE_DIFF'] = df['PRICE_RETAIL'] - df['PRICE_CURRENT']
average_price_diff = df['PRICE_DIFF'].mean()
print(f"The average price difference between retail and current prices is ${average_price_diff:.2f}.")
```
